### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
# code here
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
df.head()
```

```python
df['Year'].unique()
```

```python
Iran_series=df[df['Country'] == 'Iran'].groupby('Year')['PM2.5'].sum()
China_series=df[df['Country'] == 'China'].groupby('Year')['PM2.5'].sum()
```

```python
plt.plot(Iran_series.index, Iran_series )
plt.plot(China_series.index, China_series, linestyle='dashed' )

plt.title(" PM2.5 over the years")
plt.xlabel('Year')
plt.ylabel('PM2.5')
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'], density=True, bins=50, color='green')
plt.title("Histogram of PM10")
plt.xlabel('Bins')
plt.ylabel('Density')

plt.grid()
plt.show()
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
poland_pm = df[df['Country'] == 'Poland']
chile_pm = df[df['Country'] == 'Chile']
```

```python
plt.scatter(chile_pm['PM2.5'], chile_pm['PM10'], marker='o', color='black', label='Chile')
plt.scatter(poland_pm['PM2.5'], poland_pm['PM10'], marker='+', color='red', label='Poland')

plt.title("PM2.5 vs PM10 of Poland and Chile")
plt.xlabel('PM2.5')
plt.ylabel('PM10')

plt.legend()
plt.show()
```

### **`Problem-4:`** Draw a pie chart of top 5 most frequent countries.

```python
top_5 = df['Country'].value_counts().head(5)

plt.pie(top_5, labels=top_5.index, autopct='%0.1f%%')
plt.show()
```

### **`Problem-5:`** Draw a bar chart which represents the counts of top 5 most frequent countries.



```python
plt.bar(top_5.index, top_5)

plt.title("Count of top 5 most frequent countries")
plt.xlabel('Countries')
plt.ylabel('Count')
```

### `Problem-6`
Show a line plot of Total Profit for each month with below styling.
* Dotted Line
* Line Color Blue
* Show Legend at top left
* Circle Marker

```python
# code here
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTJh6X4_mqixWsfK9mgkllGQkKYW9Wj9kOIMGY2uYsWeS8n5np87DO-SDGQWJ1HXEnxiOVFVzYFYEcR/pub?gid=558678488&single=true&output=csv')
df.head(10)
```

```python
plt.plot(df['month_number'],df['total_profit'], linestyle='dotted', color='blue', marker='o', label='Month on Month Profit')

plt.legend(loc='upper left')

plt.title("Company Sales Profit")
plt.xlabel('Month')
plt.ylabel('Total_Profit')
plt.show()
```

### `Problem-7`
Show sales of each product in march month as pie chart.
* Show Percentage value
* Give Title "Sales in March"
* Explode ToothPaste with shadow

### `Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
q1=df.iloc[0:3]
q2=df.iloc[3:6]
q3=df.iloc[6:9]
q4=df.iloc[9:12]
```

```python
new_df=pd.DataFrame({
    'Quarter' : ['Q1','Q2','Q3','Q4', ],
    'facecream': [q1['facecream'].sum(), q2['facecream'].sum(), q3['facecream'].sum(), q4['facecream'].sum()],
    'facewash': [q1['facewash'].sum(), q2['facewash'].sum(), q3['facewash'].sum(),  q4['facewash'].sum()],
    'toothpaste': [q1['toothpaste'].sum(), q2['toothpaste'].sum(), q3['toothpaste'].sum(), q4['toothpaste'].sum()],
    'bathingsoap': [q1['bathingsoap'].sum(), q2['bathingsoap'].sum(), q3['bathingsoap'].sum(), q4['bathingsoap'].sum()],
    'shampoo': [q1['shampoo'].sum(), q2['shampoo'].sum(), q3['shampoo'].sum(),q4['shampoo'].sum()],
    'moisturizer': [q1['moisturizer'].sum(), q2['moisturizer'].sum(), q3['moisturizer'].sum(),q4['moisturizer'].sum()]

})
```

```python
new_df
```

```python
plt.bar(np.arange(new_df.shape[0])-0.5, new_df['facecream'], width=0.11, label='facecream')
plt.bar(np.arange(new_df.shape[0])-0.4, new_df['facewash'], width=0.1, label='facewash')
plt.bar(np.arange(new_df.shape[0])-0.3, new_df['toothpaste'], width=0.11, label='toothpaste')
plt.bar(np.arange(new_df.shape[0])-0.2, new_df['bathingsoap'], width=0.11, label='bathingsoap')
plt.bar(np.arange(new_df.shape[0])-0.1, new_df['shampoo'], width=0.1, label='shampoo')
plt.bar(np.arange(new_df.shape[0]), new_df['moisturizer'], width=0.1, label='moisturizer')


plt.xticks(np.arange(new_df.shape[0]), new_df['Quarter'])
plt.legend()
```
