### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


````python
l=(df['Power Station'].value_counts().head(5).index).to_list()
````

````python
l
````

````python
new_df = df[df['Power Station'].isin(l)]
````

````python
color_map={'NTPC Ltd.':1, 'NHPC':2, 'NPCIL':3, 'GMR ENERG':4, 'JSWEL':5}
colors=new_df['Power Station'].map(color_map)
````

````python
# code here
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(12,5))

ax[0].scatter(new_df['Monitored Cap.(MW)'],new_df['Total Cap. Under Maintenace (MW)'] , c=colors, cmap='summer')
ax[1].scatter(new_df['Monitored Cap.(MW)'],new_df['Actual(MU)'], c=colors, cmap='summer')

ax[0].axvline(new_df['Monitored Cap.(MW)'].mean(), c='red')
ax[1].axvline(new_df['Monitored Cap.(MW)'].mean(), c='red')

ax[0].axhline(new_df['Total Cap. Under Maintenace (MW)'].mean(), c='black')
ax[1].axhline(new_df['Actual(MU)'].mean(), c='black')

ax[0].set_xlabel('Monitored Cap')
ax[1].set_xlabel('Monitored Cap')


ax[0].set_ylabel('Total Cap. Under Maintenace')
ax[1].set_ylabel('Actual')

ax[0].set_title('Monitored Cap.(MW) vs Total Cap. Under Maintenance (MW)')
ax[1].set_title('Monitored Cap. (MW) vs Actual(MU)')

plt.show()
````

### **Problem-2:** Draw a 3D Scatter plot between "*Monitored Cap.(MW)*", "*Total Cap. Under Maintenace (MW)*" and "*Forced Maintanence(MW)*"

````python
# code here

fig=plt.figure(figsize=(10,8))

ax=plt.subplot(projection='3d')

ax.scatter3D(new_df['Monitored Cap.(MW)'], new_df['Total Cap. Under Maintenace (MW)'] , new_df['Forced Maintanence(MW)'], c=colors, cmap='viridis')

ax.set_title('Monitored Cap vs Total Cap. Under Maintenance vs Forced Maintanence')

ax.set_ylabel('Monitored Cap')
ax.set_xlabel('Total Cap. Under Maintenace')
ax.set_zlabel('Forced Maintanence(MW)')

plt.show()
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

xx, yy= np.meshgrid(x,y)
````

````python
z = np.sin(np.sqrt(xx**2 + yy**2))
````

````python
fig=plt.figure()

ax= plt.subplot(projection='3d')

ax.plot_surface(xx, yy, z, cmap='viridis')

ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')
ax.set_zlabel('Z Value')

ax.set_title("Function of z = sin(âˆšx2+y2)")

plt.show()
````

````python
fig=plt.figure()

ax= plt.subplot()

ax.contour(xx, yy, z)

ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')

ax.set_title("Function of z = sin(âˆšx2+y2)")

plt.show()
````

````python
fig=plt.figure()

ax= plt.subplot()

ax.contourf(xx, yy, z)

ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')

ax.set_title("Function of z = sin(âˆšx2+y2)")

plt.show()
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
z = np.tan(np.log2(xx**2 + yy**2))
````
